### Vâriancia 

```python
taxas_vale3 = np.array([-15.97, 13.05, -14.67, -34.53])
```

```python
media_vale3 = taxas_vale3.sum() / len(taxas_vale3)
media_vale3
```

```python
((taxas_vale3 - media_vale3)**2).sum() / len(taxas_vale3)
```

```python
variancia_vale3 = taxas_vale3.var()
variancia_vale3
```

```python
# Desvio padrao
desv_p = math.sqrt(variancia_vale3)
desv_p
```

```python
taxas_embr3 = np.array([108.69, -56.74, 44.83, 93.67])
```

```python
variancia_embr3 = taxas_embr3.var()
variancia_embr3
```

```python
# Desvio padrao embr3
desv_p_embr3 = math.sqrt(variancia_embr3)
desv_p_embr3
```

```python
# Coeficiente variação vale 3
stats.variation(taxas_vale3) * 100
```
